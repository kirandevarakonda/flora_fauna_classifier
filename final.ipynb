```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the model
model = load_model('flora_fauna_classifier.h5')

# Load an image file for prediction, resizing it to match the input size of the model
img = image.load_img("E:/flora_fauna_detection/prediction/money_plant.jpg", target_size=(150,150))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Make batch of one
img_array /= 255.0

# Make prediction
prediction = model.predict(img_array)
threshold = 0.5
if(prediction > threshold):
    predict = "flora(plant)"
    print(f"The given image is a flora(Plant)")
else:
    print(f"The given image is fauna(animal)")
    predict = "fauna(animal)"
print("Predicted class:", prediction)

plt.imshow(img)
plt.title(f"Predicted: {predict}")
plt.axis('off')  # Turn off axis labels
plt.show()
```
